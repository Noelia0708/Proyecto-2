```python
import pandas as pd
shark_attack = pd.read_excel("data/tiburones.xls")
```

```python
#ELIMINACION DE COLUMNAS INNECESARIAS PARA EL PROYECTO

shark_attack.drop('Location', axis =1, inplace=True)
shark_attack.drop('Name', axis =1, inplace=True)
shark_attack.drop('Sex', axis =1, inplace=True)
shark_attack.drop('Age', axis =1, inplace=True)
shark_attack.drop('Injury', axis =1, inplace=True)
shark_attack.drop('Time', axis =1, inplace=True)
shark_attack.drop('Source', axis =1, inplace=True)
shark_attack.drop('pdf', axis =1, inplace=True)
shark_attack.drop('href formula', axis =1, inplace=True)
shark_attack.drop('href', axis =1, inplace=True)
shark_attack.drop('Case Number', axis =1, inplace=True)
shark_attack.drop('Case Number.1', axis =1, inplace=True)
shark_attack.drop('original order', axis =1, inplace=True)
shark_attack.drop('Unnamed: 21', axis =1, inplace=True)
shark_attack.drop('Unnamed: 22', axis =1, inplace=True)
```

```python
#FILTRAR VALORES POR AÑO SUPERIORES A 1970

shark_attack = shark_attack[shark_attack['Year'] > 1970]
```

```python
#REMPLAZAR VALORES EN 'FATAL Y/N' POR 1 Y 0

shark_attack['Fatal Y/N'] = shark_attack['Fatal Y/N'].replace({'Y': 1, 'N': 0})

grouped = shark_attack.groupby('Fatal Y/N')

grouped.size()
```

```python
#UNIFICAR EL TAMAÑO DE LAS LETRAS EN COLUMNAS EXPECIFICAS

shark_attack['Country'] = shark_attack['Country'].str.lower().str.strip()
shark_attack['State'] = shark_attack['State'].str.upper().str.strip()
shark_attack['Activity'] = shark_attack['Activity'].str.lower().str.strip()
shark_attack['Type'] = shark_attack['Type'].str.lower().str.strip()
```

```python
#CAMBIAR VALORES SIMILARES A UN ÚNICO VALOR COMÚN

shark_attack = shark_attack.replace({'under investigation' : 'unknown',
                                                    'unverified' : 'unknown',
                                                    'unconfirmed' : 'unknown',
                                                    '?' : 'unknown',
                                                    'sea disaster': 'unprovoked',
                                                    'invalid' : 'unknown'})

shark_attack['State'] = shark_attack['State'].replace({'NSW': 'New South Wales'}, regex=True)
```

```python
#SUSTITUCIÓN DE VALORES NULOS

shark_attack['Country'] = shark_attack['Country'].fillna(method= 'ffill')
shark_attack['Type'] = shark_attack['Type'].fillna(method= 'ffill')
shark_attack['State'] = shark_attack['State'].fillna('unknown')
shark_attack['Activity'] = shark_attack['Activity'].fillna('unknown')
shark_attack['Fatal Y/N'] = shark_attack['Fatal Y/N'].fillna('unknown')
shark_attack['Species '] = shark_attack['Species '].fillna('unknown')
```
